```python
names, authors, rates, reviews, prices=[],[],[],[],[]
for book1 in all_books:
    names.append(book1.find(class_='p13n-sc-truncate').get_text().strip())
    authors.append(book1.find(class_='a-row a-size-small').get_text())
    rates.append(book1.find(class_='a-icon-row').get_text().strip().split('\n')[0])
    reviews.append(book1.find(class_='a-icon-row').get_text().strip().split('\n')[2])
    prices.append(book1.find(class_='a-link-normal a-text-normal').get_text())
    
```

```python
url='https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_2?ie=UTF8&pg=2'
page=requests.get(url)
soup=BeautifulSoup(page.content, 'html')
all_books=soup.find_all(class_='zg-item-immersion')
all_books
```

```python
for book1 in all_books:
    names.append(book1.find(class_='p13n-sc-truncate').get_text().strip())
    authors.append(book1.find(class_='a-row a-size-small').get_text())
    rates.append(book1.find(class_='a-icon-row').get_text().strip().split('\n')[0])
    reviews.append(book1.find(class_='a-icon-row').get_text().strip().split('\n')[2])
    prices.append(book1.find(class_='a-link-normal a-text-normal').get_text())
    
```

```python
len(names), len(authors), len(rates), len(reviews), len(prices)
```

```python
d={'Book Name':names, 'Author':authors, 'Rate':rates, 'Review':reviews, 'Price':prices}
d
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```

```python
df.to_csv('Amazon_Books.csv', index=False)
```
