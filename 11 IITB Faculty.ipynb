```python
import requests
from bs4 import BeautifulSoup
url='https://www.aero.iitb.ac.in/home/people/faculty'
page=requests.get(url)
soup=BeautifulSoup(page.content, 'html')
soup
```

```python
faculty=soup.find(class_='view-content')
faculty
```

```python
name=faculty.find_all(class_='views-field views-field-title')
name
```

```python
name1=name[1].get_text().strip()
name1
```

```python
names=[]
for i in name:
    names.append(i.get_text().strip())
names
```

```python
spec=faculty.find_all(class_='views-field views-field-field-specialization')
spec
```

```python
specs=[]
for i in spec:
    specs.append(i.get_text().strip())
specs
```

```python
room=faculty.find_all(class_='views-field views-field-field-room-no-1')
room
```

```python
rooms=[]
for i in room:
    rooms.append(i.get_text().strip())
rooms
```

```python
phone=faculty.find_all(class_='views-field views-field-field-phone-number')
phone
```

```python
phones=[]
for i in phone:
    phones.append(i.get_text().strip())
phones
```

```python
email=faculty.find_all(class_='views-field views-field-field-email')
email
```

```python
emailid=[]
for i in email:
    emailid.append(i.get_text().strip())
emailid
```

```python
d={'Name':names,'Specilization':specs,'Room':rooms,'Phone':phones,'Email':emailid}
d
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```
