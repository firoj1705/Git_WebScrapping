```python
import requests
from bs4 import BeautifulSoup
url='https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm'
page=requests.get(url)
soup=BeautifulSoup(page.content, 'html')
soup
```

```python
movies=soup.find(class_='lister-list')
movies
```

```python
titles=movies.find_all(class_='titleColumn')
titles
```

```python
title1=titles[0].get_text().strip().split('\n')
title1
```

```python
rating=movies.find_all(class_='ratingColumn imdbRating')
rating
```

```python
rate1=rating[0].get_text().strip()
rate1
```

```python
names, years, velocity=[],[],[]
for i in titles:
    names.append(i.get_text().strip().split('\n')[0])
    years.append(i.get_text().strip().split('\n')[1])
    velocity.append(i.get_text().strip().split('\n')[3])
```

```python
len(names), len(years), len(velocity)
```

```python
ratings=[]
for i in rating:
    ratings.append(i.get_text().strip())
```

```python
len(ratings)
```

```python
d={'Name':names, 'Year':years, 'Velocity':velocity,'Rate':ratings}
d
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```

```python
df.to_csv('IMDB_PopularMovies.csv')
```
