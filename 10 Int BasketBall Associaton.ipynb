```python
import requests
from bs4 import BeautifulSoup
url='https://www.basketball-reference.com/'
page=requests.get(url)
soup=BeautifulSoup(page.content, 'html')
soup
```

```python
players=soup.find(class_='images')
players
```

```python
names=players.find_all('img')
names
```

```python
name1=names[0]['data-tip']
name1
```

```python
namess=[]
for i in names:
    namess.append(i['data-tip'])
namess
```

```python
links=names[0]['src']
links
```

```python
linkss=[]
for i in names:
    linkss.append(i['src'])
linkss
```

```python
d={'Name':namess, 'Links':linkss}
d
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```
