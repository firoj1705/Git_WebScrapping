```python
import requests
from bs4 import BeautifulSoup
url='https://www.amazon.in/gp/bestsellers/watches/ref=zg_bs_nav_0'
page=requests.get(url)
soup=BeautifulSoup(page.content, 'html')
soup
```

```python
all_watches=soup.find_all(class_='zg-item-immersion')
all_watches
```

```python
watch1=all_watches[0]
watch1
```

```python
name1=watch1.find(class_='p13n-sc-truncate').get_text().strip()
name1
```

```python
ratess=watch1.find(class_='a-icon-row a-spacing-none').get_text().strip().split('\n\n')
ratess
```

```python
rate1=ratess[0]
rate1
```

```python
price1=watch1.find(class_='a-link-normal a-text-normal').get_text()
price1
```

```python
names, rates, reviews, prices=[],[],[],[]
for watch1 in all_watches:
    names.append(watch1.find(class_='p13n-sc-truncate').get_text().strip())
    rates.append(ratess[0])
    reviews.append(ratess[1])
    prices.append(watch1.find(class_='a-link-normal a-text-normal').get_text())
```
