```python
names, rates, reviews, prices=[],[],[],[]
for first in all_hk:
    names.append(first.find(class_='p13n-sc-truncate').get_text().strip())
    rates.append(first.find(class_='a-icon-row a-spacing-none').get_text().strip().split('\n\n')[0])
    reviews.append(first.find(class_='a-icon-row a-spacing-none').get_text().strip().split('\n\n')[1])
    prices.append(first.find(class_='a-link-normal a-text-normal').get_text())
```

```python
url='https://www.amazon.in/gp/bestsellers/kitchen/ref=zg_bs_pg_2?ie=UTF8&pg=2'
page=requests.get(url)
soup=BeautifulSoup(page.content, 'lxml')
all_hk=soup.find_all(class_='zg-item-immersion')
all_hk
```

```python
for first in all_hk:
    if first.find(class_='a-link-normal a-text-normal') is not None:
        names.append(first.find(class_='p13n-sc-truncate').get_text().strip())
        rates.append(first.find(class_='a-icon-row a-spacing-none').get_text().strip().split('\n\n')[0])
        reviews.append(first.find(class_='a-icon-row a-spacing-none').get_text().strip().split('\n\n')[1])
        prices.append(first.find(class_='a-link-normal a-text-normal').get_text())
    else:
        names.append(first.find(class_='p13n-sc-truncate').get_text().strip())
        rates.append(first.find(class_='a-icon-row a-spacing-none').get_text().strip().split('\n\n')[0])
        reviews.append(first.find(class_='a-icon-row a-spacing-none').get_text().strip().split('\n\n')[1])
        prices.append(None)
```

```python
len(names), len(rates), len(reviews), len(prices)
```

```python
d={'Name':names, 'Rate':rates, 'Review':reviews, 'Price':prices}
d
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```
